# Machine Learning

- Machine learning is a field of study that gives computers the ability to learn without being explicitly programmed
- Machine learning is a subset of AI, which enables the machine to automatically learn from data, improve performance from past experiences, and make predictions

***Data mining*** Applying ML techniques to dig into large amounts of data help discover patterns that were not immediately apparent

## Types of Machine Learning Systems

ML systems can be categorized based on

- Methods of learning
>- Supervised learning
>- Unsupervised learning
>- Semi-supervised learning
>- Reinforcement learning
- Online vs Batch learning
- Instance-based vs model-based learning

## Basic ML Concepts

### a. Model

A model is a representation of a system or a process that the algorithm learns from data

It's a mathematical or computational abstraction that captures the relationships, patterns and dependancies present in the data.
Examples:
- Linear regression model
- Neural network model


## 1. Supervised Machine Learning

In supervised ML algorithms the models are trained using labelled dataset and based on the training the model predicts the output.

The main goal of supervised ML techniques if to map the input variable(x) with the output variable (y)

Example of supervised ML models applications
- Risk assessment
- Fraud detection
- Spam filtering
- Price prediction

### Categories of Supervised ML

#### a. Classification

They are used to solve classification problems in which the output is categorical

They predict categories in a dataset

Example applications:
- Spam detection
- Email filtering

Example of classification algorithms

- Random Forest Algorithm
- Decision Tree Algorithm
- Logistic Regression Algorithm
- Support Vector Machine Algorithm

#### b. Regression

Regression algorithms are used to solve regression problems in which there is a linear relationship between input and output variables
These are used to predict continuous output varables

Example of regression algorithms

- Simple Linear Regression Algorithm
- Multivariate Linear Regression Algorithm
- Decision Tree Algorithm
- Lasso Regression


### Advantages and disadvantages of Supervised Learning

#### Advantages

- One has an exact idea about the classes of objects
- They are helpful in predicting the output on the basis of prior experience

#### Disadvantages

- They are not able to solve complex tasks
- May wrongly predict output if the test data is different from the training data
- Requires alot of computational time to train the algorithm

### Applications of Supervised Learning Algorithms

- Image segmentation
- Medical diagnosis
- Fraud detection
- Spam detection
- Speech recognition

## 2. Unsupervised Machine Learning

In unsupervised ML technique the training set is unlabeled.

The training data is neither classified not labelled and the model acts on that data without any supervision

The main aim of unsupervised ML algoritms is to group/categorize the unsorted data according to similarities, patterns and differences

### Categories of Unsupervised Machine Learning

#### Clustering

Involves grouping similar data points together into clusters.

The goal is to identify inherent structures or patterns in the data without prior knowledge of the classes

Example algorithms:

- K-means
- Hierarchical clustering
- Mean-shift algorithm
- DBSCAN algorithm

#### Association

It finds interesting relations among variables within a large dataset

The main of this learning algorithm is to find the dependency of one data items on another data item and map those variables accordingly so that it can generate maximum profit

The algorithm is applied in:

- Market basket analysis
- Web usage mining
- Continuous production

Some algorithms include:

- Apriori algorithm
- Eclat
- FP-growth

### Advantages and Disadvantages of Unsupervised Learning Algorithms

#### Advantages

- Can be used for complicated tasks

#### Disadvatages

- Output maybe less accurate
- More difficult to work with

### Applications of unsupervised learning

- Network analysis
- Recommendation systems
- Anomaly detection
- Singular value decomposition

## 3. Semi-Supervised Learning

These algorithms can be trained on partially labelled data.

Example:- Google photos

Most semi-supervised learning algorithms are combinations of unsupervised and supervised algorithms.

Example algorithms:-

- Deep belief networks (DBNs)
- Restricted Boltzman Machines(RBMs)

## Reinforcement Learning

The learning system, called an agent, can observe the environment, select and perform actions and get rewards in return or penalties

It learns on itself the best strategy (policy) to maximize on the rewards.

A policy defines what action the agent should choose when it is in a given situation
