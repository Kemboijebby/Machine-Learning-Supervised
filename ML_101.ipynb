# Machine Learning

- Machine learning is a field of study that gives computers the ability to learn without being explicitly programmed
- Machine learning is a subset of AI, which enables the machine to automatically learn from data, improve performance from past experiences, and make predictions

***Data mining*** Applying ML techniques to dig into large amounts of data help discover patterns that were not immediately apparent

## Types of Machine Learning Systems

ML systems can be categorized based on

- Methods of learning
>>- Supervised learning
>>- Unsupervised learning
>>- Semi-supervised learning
>>- Reinforcement learning
- Online vs Batch learning
- Instance-based vs model-based learning

## Basic ML Concepts

### a. Model

A model is a representation of a system or a process that the algorithm learns from data

It's a mathematical or computational abstraction that captures the relationships, patterns and dependancies present in the data.
Examples:
- Linear regression model
- Neural network model


## 1. Supervised Machine Learning

In supervised ML algorithms the models are trained using labelled dataset and based on the training the model predicts the output.

The main goal of supervised ML techniques if to map the input variable(x) with the output variable (y)

Example of supervised ML models applications
- Risk assessment
- Fraud detection
- Spam filtering
- Price prediction

### Categories of Supervised ML

#### a. Classification

They are used to solve classification problems in which the output is categorical

They predict categories in a dataset

Example applications:
- Spam detection
- Email filtering

Example of classification algorithms

- Random Forest Algorithm
- Decision Tree Algorithm
- Logistic Regression Algorithm
- Support Vector Machine Algorithm

#### b. Regression

Regression algorithms are used to solve regression problems in which there is a linear relationship between input and output variables
These are used to predict continuous output varables

Example of regression algorithms

- Simple Linear Regression Algorithm
- Multivariate Linear Regression Algorithm
- Decision Tree Algorithm
- Lasso Regression


### Advantages and disadvantages of Supervised Learning

#### Advantages

- One has an exact idea about the classes of objects
- They are helpful in predicting the output on the basis of prior experience

#### Disadvantages

- They are not able to solve complex tasks
- May wrongly predict output if the test data is different from the training data
- Requires alot of computational time to train the algorithm

### Applications of Supervised Learning Algorithms

- Image segmentation
- Medical diagnosis
- Fraud detection
- Spam detection
- Speech recognition
